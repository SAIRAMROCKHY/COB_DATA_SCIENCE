```python
import pandas as pd
import numpy as np
import requests
import json
```

```python
pd1=requests.get('https://data.binance.com/api/v3/ticker/24hr').json()
content = json.dumps(pd1, indent = 4, sort_keys=True)
print(content)
```

```python
df = pd.DataFrame(pd1)
df.head(10)
```

```python
df.to_csv('api_data.csv', index=True)
```

```python
df.to_csv('api_data.csv', index=False)
```
